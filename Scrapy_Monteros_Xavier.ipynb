```python
# Importa las librerías necesarias
import json
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
from docx import Document  # Biblioteca para trabajar con documentos Word
```

```python
# Función que convierte una lista de elementos en un DataFrame de pandas
def rowsToDataFrame(rows):
    df = pd.DataFrame(rows[1:], columns=rows[0])
    return df
```

```python
# Función para extraer los datos de una tabla HTML
def processTableData(tbl):
    rows = []
    for child in tbl.find('tbody').children:  # Itera los elementos internos de la tabla <tr>
        row = []
        for td in child:  # Itera los elementos internos de cada fila <td>
            try:
                item = td.text.replace('\n', '')
                if item:
                    row.append(item)
            except:
                continue
        if len(row) > 0:
            rows.append(row)
    return rows
```

```python
# Función para extraer los datos de la URL y procesar el contenido HTML
def processDataHTML(data):
    soup = bs(data['sections'][1]['text'], 'html.parser')
    tbl = soup.find_all('table')[0]
    tblRows = processTableData(tbl)
    return tblRows
```

```python
# Lee la tabla directamente desde la URL de Wikipedia
url = 'https://es.wikipedia.org/wiki/Anexo:Pa%C3%ADses_y_territorios_dependientes_por_poblaci%C3%B3n'
tables = pd.read_html(url)

# Selecciona la primera tabla (ajusta el índice si necesitas otra tabla)
df = tables[1]
```

```python
#Mostrar tabla
df
```

```python
# Función para scrapear imágenes de la página web
def scrape_images(url):
    response = requests.get(url)
    soup = bs(response.text, 'html.parser')
    images = soup.find_all('img')
    image_urls = [img['src'] for img in images if 'src' in img.attrs]
    return image_urls

print("Imágenes encontradas:")
images = scrape_images(url)
for img in images:
    print(img)
```

```python
# Función para extraer todos los enlaces de la página
def extract_links(url):
    response = requests.get(url)
    soup = bs(response.text, 'html.parser')

    # Buscar todas las etiquetas <a> con href
    links = soup.find_all('a', href=True)
    link_urls = [link['href'] for link in links]
    return link_urls

print("\nEnlaces encontrados:")
links = extract_links(url)
for link in links:
    print(link)
```

```python
# Función para buscar información específica dentro de la página
def search_information(url, search_text):
    response = requests.get(url)
    soup = bs(response.text, 'html.parser')

    # Buscar cualquier texto que contenga 'search_text'
    results = soup.find_all(text=lambda text: text and search_text.lower() in text.lower())
    return results


# Texto que deseas buscar en la página
search_text = 'territorio'

# Llamar a la función y obtener los resultados
results = search_information(url, search_text)

# Imprimir los resultados encontrados
print(f"Resultados encontrados para '{search_text}':")
if results:
    for result in results:
        print(result.strip())  # Quita espacios adicionales
else:
    print("No se encontraron coincidencias.")
```

```python
# Llamar a las funciones de scraping
images = scrape_images(url)
links = extract_links(url)
results = search_information(url, search_text)

# Crear el documento Word
doc = Document()

# Agregar la tabla al documento
doc.add_heading('Tabla', level=1)
table = doc.add_table(rows=1, cols=len(df.columns))
table.style = 'Table Grid'

# Agregar encabezados a la tabla
hdr_cells = table.rows[0].cells
for i, column in enumerate(df.columns):
    hdr_cells[i].text = str(column)

# Agregar los datos de la tabla
for index, row in df.iterrows():
    row_cells = table.add_row().cells
    for i, cell in enumerate(row):
        row_cells[i].text = str(cell)

# Agregar las imágenes al documento
doc.add_heading('Imágenes Encontradas', level=1)
if images:
    for img_url in images:
        doc.add_paragraph(img_url)
else:
    doc.add_paragraph("No se encontraron imágenes.")

# Agregar los enlaces al documento
doc.add_heading('Enlaces Encontrados', level=1)
if links:
    for link in links:
        doc.add_paragraph(link)
else:
    doc.add_paragraph("No se encontraron enlaces.")

# Agregar los resultados de búsqueda al documento
doc.add_heading(f'Resultados Encontrados para "{search_text}"', level=1)
if results:
    for result in results:
        doc.add_paragraph(result.strip())
else:
    doc.add_paragraph("No se encontraron coincidencias.")

# Guardar el documento Word
doc.save('Resultados_Scraping.docx')

print("Los datos se han guardado en el archivo 'Resultados_Scraping.docx'.")
```
